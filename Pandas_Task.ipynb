```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Salary': [50000, 60000, 55000, 80000, 52000, 58000, 120000, 51000],
    'Years_Experience': [1, 3, 2, 5, 2, 3, 10, 1]
})

stats = df.agg(['mean', 'median', 'std'])
print("Basic statistics:")
print(stats)
print()

correlation = df['Salary'].corr(df['Years_Experience'])
print(f"Correlation: {correlation:.3f}")
print()

def find_outliers_iqr(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return column[(column < lower_bound) | (column > upper_bound)]

print("Outliers in Salary:")
print(find_outliers_iqr(df['Salary']))
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve', 'Frank'],
    'Age': [15, 16, np.nan, 17, np.nan, 15],
    'Grade_Level': [9, 10, 11, 12, 9, 9]
})

print("Original with missing values:")
print(df)
print(f"\nMissing values: {df['Age'].isnull().sum()}")

df['Age'] = df.groupby('Grade_Level')['Age'].transform(
    lambda x: x.fillna(x.median())
)

print("\nAfter filling with median by Grade_Level:")
print(df)

print(f"\nMissing values after fill: {df['Age'].isnull().sum()}")
```

```python
import pandas as pd

df = pd.DataFrame({
    'Revenue': [1500, 800, 3000, 6000, 2500, 4500, 900],
    'Region': ['North', 'South', 'East', 'North', 'South', 'West', 'North'],
    'Product_Category': ['Electronics', 'Clothing', 'Food', 'Clothing',
                         'Electronics', 'Food', 'Clothing']
})

print("Original data:")
print(df)
print()

filtered = df[
    (df['Revenue'].between(1000, 5000)) &
    (df['Region'].isin(['North', 'South'])) &
    (df['Product_Category'] != 'Electronics')
]

print("Filtered results:")
print(filtered)
```

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Carol', 'David'],
    'Department': ['IT', 'HR', 'IT', 'Finance'],
    'Salary': [70000, 50000, 80000, 60000]
})

print("Original:")
print(df)
print()

df_sorted = df.sort_values(
    by=['Department', 'Salary'],
    ascending=[True, False]
)

print("Sorted:")
print(df_sorted)
print()

df_sorted = df_sorted.reset_index(drop=True)
print("With reset index:")
print(df_sorted)
```

```python
import pandas as pd

df = pd.DataFrame({
    'Department': ['IT', 'IT', 'HR', 'HR', 'Finance', 'Finance'],
    'Role': ['Dev', 'Manager', 'Recruiter', 'Manager', 'Analyst', 'Manager'],
    'Salary': [80000, 120000, 60000, 90000, 70000, 110000],
    'Years_at_Company': [2, 5, 1, 4, 3, 6]
})

print("Employee data:")
print(df)
print()

avg_salary = df.groupby('Department')['Salary'].mean()
print("1. Average Salary by Department:")
print(avg_salary)
print()

max_years = df.groupby(['Department', 'Role'])['Years_at_Company'].max()
print("2. Max Years by Department and Role:")
print(max_years)
print()

emp_count = df.groupby('Department').size()
print("3. Employee Count by Department:")
print(emp_count)
```

```python
import pandas as pd

df = pd.DataFrame({
    'Department': ['IT', 'IT', 'HR', 'HR', 'Finance', 'Finance'],
    'Salary': [80000, 120000, 60000, 90000, 70000, 110000],
    'Years_at_Company': [2, 5, 1, 4, 3, 6]
})

result = df.groupby('Department').agg({
    'Salary': ['sum', 'mean', 'count'],
    'Years_at_Company': 'max'
})

result.columns = ['Total_Salary', 'Avg_Salary', 'Emp_Count', 'Max_Years']
result = result.reset_index()

print(result)
```
