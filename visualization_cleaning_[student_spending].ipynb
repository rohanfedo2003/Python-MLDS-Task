```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x="gender", data=df1_copy)
plt.title("Gender Distribution")
plt.show()
```

```python
sns.countplot(x="year_in_school", data=df1_copy)
plt.title("Students by Year")
plt.show()
```

```python
sns.histplot(df1_copy["age"], bins=20)
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(df1_copy["monthly_income"], bins=20)
plt.title("Monthly Income")
plt.show()
```

```python
sns.scatterplot(x="food", y="housing", data=df1_copy)
plt.title("Food vs Housing Spending")
plt.show()
```

```python
sns.boxplot(x="gender", y="food", data=df1_copy)
plt.title("Food Spending by Gender")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df1_copy.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x="monthly_income", y="entertainment", data=df1_copy)

plt.title("Monthly Income vs Entertainment Spending")
plt.xlabel("Monthly Income")
plt.ylabel("Entertainment Spending")

plt.show()
```
