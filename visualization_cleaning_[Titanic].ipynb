```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("DATASET/Titanic-Dataset.csv")
df
```

```python
df1_copy=df.copy()
df1_copy
```

```python
df1_copy.columns
```

```python
df1_copy.info()
```

```python
df1_copy.isnull().sum()
```

```python
df1_copy[df1_copy["Age"].isnull()]
```

```python
# df1_copy["Age"].fillna(df1_copy['Age'].mean(),inplace= True)

df1_copy["Age"] = df1_copy["Age"].fillna(df1_copy["Age"].mean())
```

```python
df1_copy.isnull().sum()
```

```python
df1_copy = df1_copy.drop("Cabin", axis=1)
```

```python
df1_copy.columns
```

```python
df1_copy.isnull().sum()
```

```python
df1_copy["Embarked"].mode()[0]
```

```python
df1_copy['Embarked']=df1_copy['Embarked'].fillna(df1_copy["Embarked"].mode()[0])
```

```python
df1_copy.isnull().sum()
```

```python
df1_copy.shape
```

```python
df1_copy.info()
```

```python
sns.countplot(x="Survived", data=df1_copy)
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x="Sex", hue="Survived", data=df1_copy)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(x="Pclass", hue="Survived", data=df1_copy)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
df1_copy["Age"].max()
```

```python
sns.histplot(df1_copy["Age"], bins=20)
plt.title("Age Distribution")
plt.show()
```

```python
df1_copy["Fare"].max()
```

```python
sns.histplot(df1_copy["Fare"], bins=20)
plt.title("Fare Distribution")
plt.show()
```

```python
sns.boxplot(x="Survived", y="Age", data=df1_copy)
plt.title("Age vs Survival")
plt.show()
```
