## Import Libraries

```python

import pandas as pd
import numpy as np
```

## 1. Select rows where Age >= 31

```python

data = {
'Name':['Teodosija','Sutton','Taneli','Ravshan','Ross'],
'Age':[26,32,25,31,28],
'Salary':[50000,60000,45000,70000,55000]
}

df = pd.DataFrame(data)

df[df['Age'] >= 31]
```

## 2. Age > 25 AND Salary > 50000

```python

df[(df['Age'] > 25) & (df['Salary'] > 50000)]
```

## 3. Remove Duplicate Rows

```python

data = {
'Name':['Ross','Bob','Ross','Geoffrey','Bob'],
'Age':[25,30,25,22,30],
'Salary':[50000,60000,50000,45000,60000]
}

df = pd.DataFrame(data)

df.drop_duplicates()
```

```python

array_1d = np.array([1,2,3,4,5,6,7,8])

array_2d = array_1d.reshape(2,4)

array_2d
```

## Data Selection with loc

```python

df = pd.DataFrame({
'Product':['Laptop','Mouse','Keyboard','Monitor','Headphones'],
'Price':[999,25,75,300,150],
'Stock':[15,150,80,45,60],
'Category':['Electronics','Accessories','Accessories','Electronics','Audio']
}, index=['P001','P002','P003','P004','P005'])

df
```

### 1. Select row P003

```python

df.loc['P003']
```

### 2. Rows P002 to P004 with Product and Price

```python

df.loc['P002':'P004', ['Product','Price']]
```

### 3. Category Electronics

```python

df.loc[df['Category'] == 'Electronics']
```

## iloc Examples

### First 3 rows

```python

df.iloc[:3]
```

### Rows at position 1 and 3

```python

df.iloc[[1,3]]
```

### Last 2 rows and first 2 columns

## Filtering with Conditions

```python

df = pd.DataFrame({
'Name':['John','Sarah','Mike','Emma','Chris','Anna'],
'Department':['IT','HR','IT','Sales','Sales','HR'],
'Salary':[75000,65000,80000,55000,58000,62000],
'Years':[3,5,7,2,4,6],
'Remote':[True,False,True,False,True,False]
})

df
```

### Employees in IT

```python

df[df['Department'] == 'IT']
```

### Salary above 60000

```python

df[df['Salary'] > 60000]
```

### Remote workers

```python

df[df['Remote'] == True]
```

### Experience >= 5 years

```python

df[df['Years'] >= 5]
```

## Mixed loc and iloc

### Salary of 3rd row

## Multiple Conditions (AND)

```python

df = pd.DataFrame({
'Name':['John','Sarah','Mike','Emma','Chris','Anna'],
'Department':['IT','HR','IT','Sales','Sales','HR'],
'Salary':[75000,65000,80000,55000,58000,62000],
'Years':[3,5,7,2,4,6],
'Remote':[True,False,True,False,True,False]
})

df[(df['Department']=='IT') & (df['Remote']==True)]
```

```python

df[(df['Salary'].between(60000,75000)) & (df['Years'] > 3)]
```

```python

df[(df['Department']!='HR') & (df['Salary'] > 60000)]
```

## Multiple Conditions (OR)

```python

df[(df['Department']=='IT') | (df['Remote']==True)]
```

```python

df[(df['Salary'] < 60000) | (df['Salary'] > 70000)]
```

```python

df[(df['Department']=='Sales') | (df['Years'] < 3)]
```

## Complex Real-world Filtering

```python

df = pd.DataFrame({
'Date': pd.date_range('2024-01-01', periods=10),
'Region':['North','South','East','West','North','South','East','West','North','South'],
'Product':['A','B','A','C','B','A','C','B','A','C'],
'Quantity':[10,15,8,20,12,25,18,22,14,30],
'Revenue':[1000,1500,800,2000,1200,2500,1800,2200,1400,3000]
})
```

```python

df[(df['Region'].isin(['North','South'])) & (df['Revenue'] > 1200)]
```

```python

df[(df['Product'].isin(['A','B'])) & (df['Quantity'].between(10,20))]
```

```python

top50 = df['Revenue'].quantile(0.5)

df[(df['Region']!='West') & (df['Revenue'] >= top50)]
```
