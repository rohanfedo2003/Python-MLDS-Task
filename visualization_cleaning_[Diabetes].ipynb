```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("DATASET/diabetes.csv")

df.head()
```

```python
df1_copy=df.copy()
df1_copy
```

```python
df1_copy.columns
```

```python
df1_copy.shape
```

```python
df1_copy.info()
```

```python
df1_copy.describe()
```

```python
cols = ["Glucose","BloodPressure","SkinThickness","Insulin","BMI"]

df1_copy[cols] = df1_copy[cols].replace(0, np.nan)
```

```python
df1_copy.isnull().sum()
```

```python
df1_copy.fillna(df1_copy.median(), inplace=True)
```

```python
df1_copy.isnull().sum()
```

```python
sns.countplot(x="Outcome", data=df1_copy)

plt.title("Diabetes Count")
plt.show()
```

```python
sns.histplot(df1_copy["Age"], bins=20)

plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(df1_copy["Glucose"], bins=20)

plt.title("Glucose Distribution")
plt.show()
```

```python
sns.boxplot(x="Outcome", y="Age", data=df1_copy)

plt.title("Age vs Diabetes")
plt.show()

# Numerical vs Category
```

```python
sns.scatterplot(x="Age", y="BMI", data=df1_copy)
plt.show()
# Relationship Between Two Numbers
```
