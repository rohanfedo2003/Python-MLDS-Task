```python
password = input("Enter password: ")

if len(password) < 8:
    print("Weak password (too short)")
elif (any(c.isupper() for c in password) and
      any(c.islower() for c in password) and
      any(c.isdigit() for c in password)):
    print("Strong password!")
else:
    print("Weak password")
```

```python
def mask_email(email):
    # Check if email is valid
    if "@" not in email:
        return "Invalid email format"
    
    username, domain = email.split("@", 1)

    
    if len(username) <= 2:
        return email

    
    masked_username = username[0] + "*" * (len(username) - 2) + username[-1]
    
    return masked_username + "@" + domain


def mask_phone(phone):
    
    digits = ''.join(char for char in phone if char.isdigit())

    
    if len(digits) < 4:
        return phone

    
    masked = "*" * (len(digits) - 4) + digits[-4:]

    return masked



email = input("Enter email: ")
phone = input("Enter phone number: ")

print("Masked email:", mask_email(email))
print("Masked phone:", mask_phone(phone))
```

```python
text = input("Enter text: ")

total_chars = len(text)
total_chars_no_space = len(text.replace(" ", ""))
words = text.split()
total_words = len(words)

avg_word_length = sum(len(word) for word in words) / total_words if total_words > 0 else 0

from collections import Counter
most_common_char = Counter(text).most_common(1)

sentences = text.count(".") + text.count("!") + text.count("?")

print("Total characters:", total_chars)
print("Characters without spaces:", total_chars_no_space)
print("Total words:", total_words)
print("Average word length:", round(avg_word_length, 2))
print("Most common character:", most_common_char)
print("Number of sentences:", sentences)
```
