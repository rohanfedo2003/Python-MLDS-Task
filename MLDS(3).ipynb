```python
words = ["cat", "elephant", "dog"]

word_lengths = {}

for word in words:
    word_lengths[word] = len(word)

print(word_lengths)
```

```python
def count_characters(text):
    char_count = {}

    for char in text:
        if char != " ":   
            if char in char_count:
                char_count[char] += 1
            else:
                char_count[char] = 1

    return char_count


print(count_characters("hello world"))
```

```python
from collections import Counter

def count(name):
    return dict(Counter(name.replace(" ", "")))

print(count("hello world"))
```
